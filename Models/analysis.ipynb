```python
#connect to database
engine = create_engine(f"postgres://postgres:postgres@localhost:5432/adoption_db")
conn = engine.connect()
```

Pet Finder Data

```python
# query data
petfinder= pd.read_sql("select * FROM animal WHERE page is not null AND test_train='train'", conn)
petfinder.dropna()
```

Kaggle Data

```python
# query data
kaggle_df= pd.read_sql("select * FROM animal WHERE page is null AND test_train='train'", conn)
kaggle_df.head()
```

```python
X = petfinder[["type", "age", "breed1", "breed2", "gender", "color1",
      "color2", "color3", "maturity_size", "furlength", "vaccinated", 
      "dewormed", "sterilized", "health", "fee","photo_amt","video_amt"]]
y = petfinder[["adoption_speed"]]
# X_test = test_df[["type", "age", "breed1", "breed2", "gender", "color1",
#       "color2", "color3", "maturity_size", "furlength", "vaccinated", 
#       "dewormed", "sterilized", "health", "fee","photo_amt","video_amt"]]
# X
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.layers import Dense

X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=0)


X_scaler = MinMaxScaler().fit(X)
X_train_scaled = X_scaler.transform(X)
X_test_scaled = X_scaler.transform(X_test)
                                   
y_train_categorical = to_categorical(y)
y_test_categorical = to_categorical(y_test,5)
```

```python
y_test_categorical
```

```python
model=Sequential()
```

```python
model.add(Dense(units=200, activation='relu', input_dim=17))
model.add(Dense(units=200, activation="relu"))
model.add(Dense(units=200, activation="relu"))
model.add(Dense(units=200, activation="relu"))
model.add(Dense(units=5, activation="softmax")),
```

```python

model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])

model.fit(X_train_scaled, y_train_categorical, epochs=250, shuffle=True, verbose=2)
```

```python
model.predict(X_test_scaled)
```

```python
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(
    f"PetFinder Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```
